```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generowanie danych
np.random.seed(42)  # aby wyniki były powtarzalne
kategorie = ["Żywność", "Mieszkanie", "Transport", "Rozrywka", "Inne"]
miesiace = [
    "Styczeń", "Luty", "Marzec", "Kwiecień", "Maj", "Czerwiec",
    "Lipiec", "Sierpień", "Wrzesień", "Październik", "Listopad", "Grudzień"
]

# wydatki losowe: 1000–4000 dla mieszkania, 500–2000 dla żywności, itd.
dane = {
    "Żywność": np.random.randint(500, 2000, size=12),
    "Mieszkanie": np.random.randint(1000, 4000, size=12),
    "Transport": np.random.randint(300, 1500, size=12),
    "Rozrywka": np.random.randint(200, 1200, size=12),
    "Inne": np.random.randint(100, 800, size=12)
}

# 2. Tworzenie DataFrame
df = pd.DataFrame(dane, index=miesiace)

# 3. Analiza danych
# całkowite wydatki na każdą kategorię
calkowite_wydatki = df.sum()

# średnie wydatki miesięczne na kategorię
srednie_wydatki = df.mean()

# miesiąc z najwyższymi wydatkami ogółem
df["Suma"] = df.sum(axis=1)
miesiac_max = df["Suma"].idxmax()
max_wartosc = df["Suma"].max()

print("Całkowite wydatki w roku:")
print(calkowite_wydatki, "\n")
print("Średnie wydatki miesięczne:")
print(srednie_wydatki, "\n")
print(f"Miesiąc z najwyższymi wydatkami: {miesiac_max} ({max_wartosc} zł)\n")

# 4. Wizualizacja danych
df[kategorie].plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Wydatki miesięczne wg kategorii (2023)")
plt.ylabel("Kwota (PLN)")
plt.xlabel("Miesiąc")
plt.legend(title="Kategorie")
plt.tight_layout()
plt.show()
```
